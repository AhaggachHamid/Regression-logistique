```r
education.mean = mean(data$education, na.rm = TRUE)
education.mean1 = ifelse(education.mean > 1.5, 2, FALSE)
data[is.na(data$education) == TRUE, "education"] = education.mean1

cigsPerDay.median = median(data$cigsPerDay, na.rm = TRUE)
data[is.na(data$cigsPerDay) == TRUE, "cigsPerDay"] =cigsPerDay.median

BPMeds.null = 0
data[is.na(data$BPMeds) == TRUE, "BPMeds"] = BPMeds.null

totChol.median = median(data$totChol, na.rm = TRUE)
data[is.na(data$totChol) == TRUE, "totChol"] =totChol.median

BMI.median = median(data$BMI, na.rm = TRUE)
data[is.na(data$BMI)== TRUE, "BMI"] =BMI.median

heartRate.median = median(data$heartRate, na.rm = TRUE)
data[is.na(data$heartRate) == TRUE, "heartRate"] = heartRate.median

glucose.median = median(data$glucose, na.rm = TRUE)
data[is.na(data$glucose) == TRUE, "glucose"] =glucose.median
```

```r
data$education1 = ifelse(data$education == 1,1,0)
data$education2 = ifelse(data$education == 2,1,0)
data$education3 = ifelse(data$education == 3,1,0)
data$education4 = ifelse(data$education == 4,1,0)
```

```r
glm.fit =glm(TenYearCHD ~., data = data, family = binomial)
```

```r
summary(glm.fit)
```

```r
glm.fit1=glm(TenYearCHD ~ male+age+cigsPerDay+prevalentStroke+prevalentHyp+totChol+sysBP+glucose, data = data, family = binomial)
```

```r
summary(glm.fit1)
```

```r
exp(coefficients(glm.fit1))
```

```r
step(glm.fit,trace=TRUE, direction="backward")
```

```r
glm.probs = predict(glm.fit1,type = "response")
pb=mean(glm.probs)
glm.pred =ifelse(glm.probs > pb, "1", "0")

# glm.pred[0:20]
```

```r
predict=predict(glm.fit1,data.frame(male=1 ,age=60,cigsPerDay=45 , prevalentStroke =1 ,
    prevalentHyp=1, totChol=200, sysBP=107,glucose=40),interval="prediction",level=0.95)
glm.pred =ifelse(predict > pb, "1", "0")
glm.pred
```
